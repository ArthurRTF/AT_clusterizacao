```python
# Questão 12
import idx2numpy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

X_train = idx2numpy.convert_from_file("train-images.idx3-ubyte")
y_train = idx2numpy.convert_from_file("train-labels.idx1-ubyte")
X_test  = idx2numpy.convert_from_file("t10k-images.idx3-ubyte")
y_test  = idx2numpy.convert_from_file("t10k-labels.idx1-ubyte")

X_achatado = X_train.reshape((X_train.shape[0], -1))

X_sample = X_achatado[:5000]
y_sample = y_train[:5000]

std      = StandardScaler()
X_scaled = std.fit_transform(X_sample)

tsne    = TSNE(n_components=2, random_state=42)
X_tsne  = tsne.fit_transform(X_scaled)

plt.figure(figsize=(10,8))
sns.scatterplot(x=X_tsne[:,0], y=X_tsne[:,1], hue=y_sample, palette="tab10")
plt.title("MNIST com t-SNE")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")
plt.legend(title="Dígito", loc='upper right')
plt.show()
```

O MNIST possui muitas dimensões, tornando difícil uma visualização direta. O t-SNE permitiu reduzir os dados para 2D, revelando agrupamentos de dígitos semelhantes: alguns números, como 0, 1, 6 e 7, ficaram mais separados, enquanto outros, como 4 e 9 ou o 3 e 5, apresentaram uma sobreposição. Isso aí já mostra tanto a utilidade do t-SNE para interpretar dados de alta dimensão quanto os desafios de separar classes muito parecidas.
